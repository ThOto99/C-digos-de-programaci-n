```python
import pandas as pd
from google.colab import files  # Para subir archivos en Google Colab
```

```python
# Subir archivo CSV desde la PC
display("Por favor, sube tu archivo CSV")
uploaded = files.upload()
```

```python
# Obtener el nombre del archivo subido
file_path = list(uploaded.keys())[0]
print(f"Archivo cargado: {file_path}")
```

```python
# Leer el archivo CSV en un DataFrame
df = pd.read_csv(file_path)
```

```python
# Ver las primeras filas del archivo para entender su estructura
print("Primeras filas del DataFrame:")
df.head()
```

```python

# Normalizar el nombre de la columna para SMILES
possible_smiles_columns = ["SMILES", "Smiles", "SMILES ", "Smiles "]
smiles_column = next((col for col in possible_smiles_columns if col in df.columns), None)
if not smiles_column:
    raise ValueError("No se encontró una columna de SMILES en el archivo CSV.")

df[smiles_column] = df[smiles_column].str.strip()  # Eliminar espacios alrededor de los SMILES
```

```python
# Verificar si la columna de actividad -log(M) existe
activity_column = "-.log(M)"
if activity_column not in df.columns:
    raise ValueError(f"No se encontró la columna {activity_column} en el archivo CSV.")
```

```python
# Convertir la columna -log(M) a numérica y manejar errores
df[activity_column] = pd.to_numeric(df[activity_column], errors='coerce')
```

```python
# Identificar SMILES con actividad NaN o 0
smiles_con_actividad_nula = df[df[activity_column].isna() | (df[activity_column] == 0)]
```

```python
# Filtrar el DataFrame eliminando duplicados, pero manteniendo los de mayor actividad
# y reincorporando los SMILES con actividad NaN o 0
df_filtrado = df.dropna(subset=[activity_column])  # Eliminar NaN temporalmente para ordenación
df_filtrado = df_filtrado.sort_values(by=[smiles_column, activity_column], ascending=[True, False])
df_filtrado = df_filtrado.drop_duplicates(subset=[smiles_column], keep='first')

df_final = pd.concat([df_filtrado, smiles_con_actividad_nula]).drop_duplicates(subset=[smiles_column], keep='first')
```

```python
# Ver las primeras filas después del procesamiento
print("Primeras filas del DataFrame limpio:")
print(df_final.head())
```

```python
# Guardar el nuevo DataFrame en un archivo CSV
output_file = "pMic_limpio.csv"
df_final.to_csv(output_file, index=False)
print(f"Archivo procesado y guardado como: {output_file}")
```
