```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import SaltRemover
from google.colab import files
```

```python
# Cargar el archivo CSV desde la computadora
uploaded = files.upload()
file_path = list(uploaded.keys())[0]  # Obtener el nombre del archivo subido
```

```python
# Leer el archivo CSV en un DataFrame
df = pd.read_csv(file_path)
```

```python
# Normalizar el nombre de la columna para SMILES
possible_smiles_columns = ["SMILES", "Smiles", "SMILES ", "Smiles "]
smiles_column = next((col for col in possible_smiles_columns if col in df.columns), None)
if not smiles_column:
    raise ValueError("No se encontró una columna de SMILES en el archivo CSV.")
```

```python
# Eliminar espacios en los SMILES y manejar errores
def validate_smiles(value):
    """Convertir el valor a string y validar que sea un SMILES."""
    try:
        value = str(value).strip()  # Convertir a string y eliminar espacios
        mol = Chem.MolFromSmiles(value)  # Validar si puede ser convertido a Mol
        if mol is not None:
            return value
    except Exception as e:
        pass
    return None  # Retornar None si no es válido
```

```python
# Aplicar validación
df[smiles_column] = df[smiles_column].apply(validate_smiles)
```

```python
# Filtrar filas con SMILES inválidos
df = df.dropna(subset=[smiles_column]).reset_index(drop=True)
```

```python
# Función para remover sales y iones metálicos
def remover_sales_y_iones(mol):
    remover = SaltRemover.SaltRemover()
    mol_limpio = remover.StripMol(mol, dontRemoveEverything=True)
    return mol_limpio
```

```python
# Función para limpiar SMILES
def limpiar_estructura(smiles):
    try:
        mol = Chem.MolFromSmiles(smiles)
        if not mol:
            return None
        Chem.SanitizeMol(mol)
        mol_limpio = remover_sales_y_iones(mol)
        return Chem.MolToSmiles(mol_limpio)
    except:
        return None

df["Smiles limpios"] = df[smiles_column].apply(limpiar_estructura)
```

```python
# Guardar el resultado en un archivo CSV
df.to_csv("Smiles_limpios.csv", index=False)
print("Archivo procesado y guardado como: Smiles_limpios.csv")
```
